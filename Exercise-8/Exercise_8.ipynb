# **Matplotlib**

```python
import matplotlib
```

Matplotlib Pyplot = For plotting graphs using matplotlib
Numpy = Numerical python

```python
import matplotlib.pyplot as plt
import numpy as np

#Coordinates
x = np.array([0,6])
y = np.array([0,250])

plt.plot(x, y) #or you could only plot either x or y only
plt.show() #shows the plotted graph
```

Plotting without line

```python
plt.plot(x, y, 'o') #'o' means rings, thus plotting only the markers
plt.show()
```

Plotting array

```python
arr1 = np.array([3,8,1,10])
plt.plot(arr1, marker='o') #Adding marker means the plot will make lines along with the rings
plt.show()
```

```python
plt.plot(arr1, marker='x') #'x' means x marks instead of rings
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=20) # 'o:b' means having ring markers and broken lines, and ms increases the size of the markers
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=10, mec='r') # Adding mec adds colored borders to the markers. In this case, 'r' means red
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=25, mec='green',mfc='red') #mfc changes the color of the markers
plt.show()
```

```python
plt.plot(arr1, linestyle='dashed') #linestyle also changes the linestyle duhh
plt.show()
```

```python
plt.plot(arr1, linestyle='--') #another way of saying dashed
plt.show()
```

Adding labels and titles

```python
xarr = np.array([1,2,6,8])
yarr = np.array([3,8,1,10])

plt.plot(xarr, yarr)

plt.title("Title of the whole graph")
plt.xlabel("Label of x axis")
plt.ylabel("Label of y axis")
plt.show()
```

Adding more shenanigans

```python
plt.plot(xarr, yarr)

plt.title("Title of the whole graph", loc='left') #prints the text in left alignment
plt.xlabel("Label of x axis")
plt.ylabel("Label of y axis")
plt.grid() #adds grid for both x and y axis
plt.show()
```

```python
plt.plot(xarr, yarr)

plt.title("Title of the whole graph", loc='left')
plt.xlabel("Label of x axis")
plt.ylabel("Label of y axis")
plt.grid(axis='y') #adds grid for either y or x
plt.show()
```

```python
plt.plot(xarr, yarr)

plt.title("Title of the whole graph", loc='left')
plt.xlabel("Label of x axis")
plt.ylabel("Label of y axis")
plt.grid(axis='x',linestyle='--',linewidth=0.50) #changes how the grid looks like
plt.show()
```

Subplots and multiple plots

```python
#first plot
xarr1 = np.array([1,2,6,8])
yarr1 = np.array([3,8,1,10])

plt.subplot(1,2,1) #Changes the size or dimention of the graph
plt.plot(xarr1,yarr1)

#second plot
xarr2 = np.array([1,2,6,8])
yarr2 = np.array([10,15,20,25])

plt.subplot(1,2,2) #Changes the size or dimention of the graph
plt.plot(xarr2,yarr2)

plt.show()
```

Scatterplot

Bar chart

```python
variables = np.array(['Label1', 'Label2', 'Label3', 'Label4', 'Label5'])
frequency = np.array([90, 85, 50, 65, 120])

plt.bar(variables,frequency,color='red') #changes the color of the graph
plt.show()
```

Pie chart

```python
percentage = np.array([25, 10, 15, 45, 5])
chart = np.array(["Label1 = 25%", "Label2 = 10%", "Label3 = 15%", "Label4 = 45%", "Label5 = 5%"])

plt.pie(percentage, labels=chart)
plt.show()
```

Add legend

```python
plt.pie(percentage, labels=chart)
plt.legend(labels=chart) #adds legend
plt.show()
```

Making dataframe

```python
import pandas

dat = pandas.read_csv('Health_Sleep_Statistics.csv')
df_dat = pandas.DataFrame(dat)
print(df_dat)
```

Question 1: How many respondents have sleep disorders?

```python
import matplotlib.pyplot as plt

Category = ['Yes', 'No']
sleepD = (df_dat["Sleep Disorders"].values.tolist())
sleepVal = [sleepD.count('yes'), sleepD.count('no')]
plt.bar(Category, sleepVal)
plt.title("Number of respondents with sleep disorders")
plt.xlabel("With sleep disorders")
plt.ylabel("Frequency")
plt.show()
```

Question 2: What are the dietary habits of the respondents?

```python
categ = ['Healthy', 'Medium', 'Unhealthy']
habit = (df_dat["Dietary Habits"].values.tolist())
habitPercent = [habit.count('healthy'), habit.count('medium'), habit.count('unhealthy')]
plt.pie(habitPercent, labels=categ, autopct='%1.1f%%') #autopct shows the percentage of each categories
plt.title("Dietary habits of the respondents")
plt.show()
```

Question 3: What are the sleep quality of all the respondents?

```python
SleepQ = df_dat["Sleep Quality"].values
plt.hist(SleepQ)
plt.title("SLeep quality of every respondent")
plt.xlabel("Respondent no.")
plt.ylabel("Sleep Quality")
plt.show()
```

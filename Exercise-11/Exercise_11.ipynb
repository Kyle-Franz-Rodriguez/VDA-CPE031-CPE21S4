```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read the file
xy = pd.read_csv('Sample-Problem.csv')
xy
```

# Use linear model and interpret your data.

```python
#Generate data
x = xy['x']
y = xy['y']

#Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate the predicted values of y using the regression line equation
prediction = slope * x + intercept

#Plot the data points and the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, prediction, color='red', label='Regression Line')
plt.show()
```

The graph shows the relationship of the x and y variable via regression line. Based from the visualization, it can be concluded that the x and y have a strong negative linear correlation.

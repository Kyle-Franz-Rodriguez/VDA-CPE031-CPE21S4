```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart_2022_with_nans.csv')
df = pd.DataFrame(data)
print(df)
```

# Question 1. How healthy are the data subjects based on their habit?

```python
categorybar = ["Alcohol drinkers", "HIV testing", 'FluVaxLast12', 'PneumoVaxEver', "COVID Positive", 'HighRiskLastYr']
alcohol = df['AlcoholDrinkers'].values.tolist()
HIV = df['HIVTesting'].values.tolist()
Flu = df['FluVaxLast12'].values.tolist()
Pneu = df['PneumoVaxEver'].values.tolist()
COVID = df['CovidPos'].values.tolist()
HighRisk = df['HighRiskLastYear'].values.tolist()
alcoholno = alcohol.count('Yes')
HIVno = HIV.count('Yes')
Fluno = Flu.count('Yes')
Pneuno = Pneu.count('Yes')
COVIDno = COVID.count('Yes')
HighRiskno = HighRisk.count('Yes')
categorycount = [alcoholno, HIVno, Fluno, Pneuno, COVIDno, HighRiskno]
plt.bar(categorybar, categorycount)
plt.xticks(rotation='vertical')
plt.show()
```

Based on the given bar chart, I have observed the following: almost half of the population are alcohol drinkers, about 125000 tested for HIV, almost half are flu vaccinated last 12 months, about 150000 are pneumonia vaccinated, more than 100000 are COVID-19 postive, and less than 25000 are at a high risk of heart disease last year. With this, I can say that even though more than half of the population are alcohol drinkers, 1/4 of the population are COVID positive, and a few are at risk, more than half of the test subjects have a reasonable healthy habit.

# Question 2. How well are the general health of the data subjects?

```python
GenH = df['GeneralHealth'].value_counts()
healthrate = GenH.index
plt.pie(GenH, labels=healthrate, autopct='%1.1f%%')
plt.show()
```

Based on the pie chart created, only 4.4% have a poor general health. With the inclusion of the aforementioned bar chart, it is safe to say that even though at least half showed an unhealthy habit, it is safe to say that the 81.9% of the test subjects, who have a range of fair to excellent, are keeping a healthy life that drinking alcohol and being COVID positive doesn't affect their health by a large margin.
